## Overview

Edulance is an AI-powered student freelancing and academic collaboration platform that helps students connect, collaborate, and grow. Students can post academic tasks, find skilled peers, communicate securely, and receive intelligent AI assistance powered by Gemma.

The platform is designed to make learning more accessible while creating opportunities for students to earn through their skills.

# Problem Statement

Many students struggle to:

- Find trustworthy academic help
- Connect with skilled peers
- Get quick explanations while studying
- Earn money from their knowledge
- Access AI assistance designed for education

Most existing platforms are either expensive, not student-focused, or lack intelligent academic support.

# Our Solution

Edulance provides one platform where students can:

- Post academic tasks
- Accept freelance work
- Chat securely
- Share knowledge
- Build their reputation
- Receive AI-powered assistance using Gemma

The application encourages peer learning while making education more affordable and accessible.

# Key Features

- AI Academic Assistant powered by Gemma
- Student Freelancing Marketplace
- Task Posting System
- Secure Authentication
- Real-time Chat
- Profile & Ratings
- Skill Discovery
- Mobile Application (Android)
- Responsive Landing Website

# How Gemma is Used

Gemma powers the intelligent assistant inside Edulance.

The assistant helps students by:

- Improving task descriptions
- Explaining academic concepts
- Answering study questions
- Assisting with programming problems
- Providing learning guidance
- Helping students communicate more effectively

This enables students to complete academic work faster while improving learning quality.

# System Architecture

```
                    Users
                      │
                      ▼
             React Native (Expo)
                      │
      ┌───────────────┼────────────────┐
      ▼               ▼                ▼
 Firebase Auth   Cloud Firestore   Cloudinary
      │               │                │
      └───────────────┼────────────────┘
                      │
                      ▼
             Gemma AI Assistant
                      │
                      ▼
         Intelligent Student Support
```

The application securely manages authentication, stores user and task data in Firebase, uploads files through Cloudinary, and uses Gemma to provide intelligent educational assistance.

# Workflow

1. Student signs in.
2. Student creates a profile.
3. User posts an academic task.
4. Other students browse available tasks.
5. Interested students apply.
6. Users communicate through chat.
7. Gemma assists users with academic guidance.
8. Tasks are completed securely.
9. Students learn, collaborate and build experience.
